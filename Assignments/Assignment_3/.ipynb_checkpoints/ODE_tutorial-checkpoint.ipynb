```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
```

Numerical solution of 

$ \frac{dy(t)}{dt}=-ky(t) $

## Tidy Equation: Everything known

```python
def model(y,t):
    k = 0.3
    dydt = -k*y
    return dydt
```

```python
# Initial conditions
y0 = 5

# Time points
t = np.linspace(0, 20, 101)

# Solve ODE
y = odeint(model, y0, t)
```

```python
plt.plot(t, y)
plt.xlabel('time')
plt.ylabel('y(t)')
plt.show()
```

## Unkown Parameter

```python
def model(y,t,k):
    dydt = -k*y
    return dydt
```

```python
k=0.1
y1 = odeint(model, y0, t, args=(k,))
k=0.2
y2 = odeint(model, y0, t, args=(k,))
k=0.3
y3 = odeint(model, y0, t, args=(k,))
```

```python
plt.plot(t, y1)
plt.plot(t, y2)
plt.plot(t, y3)
plt.xlabel('time')
plt.ylabel('y(t)')
plt.show()
```

## Excercises

### Problem 1

$ \frac{dy(t)}{dt} = -y(t) + 1 $

```python
def model(y, t):
    dydt = -y + 1
    return dydt

t = np.linspace(0, 5, 51)

y = odeint(model, 0, t)

plt.plot(t, y)
plt.xlabel('time')
plt.ylabel('y(t)')
plt.show()
```

### Problem 2

$ 5\frac{dy(t)}{dt} = -y(t) + u(t) $

$ y(0) = 1 $

```python
def model(y, t):
    if (t<10.0):
        u = 0.0
    else: 
        u = 2.0
    dydt = -1/5*y + 1/5*u
    return dydt

y0 = 1

t = np.linspace(0, 35, 51)

y = odeint(model, y0, t)

plt.plot(t, y)
plt.xlabel('time')
plt.ylabel('y(t)')
plt.show()
```

### Problem 3

$ \frac{dx(t)}{dt} = 3e^{-t} $

$ \frac{dy(t)}{dt} = 3 - y(t) $

$ y(0) = x(0) = 0 $

```python
def model(initial, t):
    x = initial[0]
    y = initial[1]
    dxdt = 3.0 * np.exp(-t)
    dydt = 3 - y
    return dydt

y0 = 1

t = np.linspace(0, 35, 51)

y = odeint(model, y0, t)

plt.plot(t, y)
plt.xlabel('time')
plt.ylabel('y(t)')
plt.show()
```
